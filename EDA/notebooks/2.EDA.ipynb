## Prueba de normalidad 

### 1. Prom_ganancias

```python
df2= df['prom_ganancias'].dropna()
df2.isna().sum()
```

```python
#  Prueba de D'Agostino
stat, p_valor = normaltest(df2)

# Mostrar los resultados
print("Estadístico de prueba:", stat)
print("P-valor:", p_valor)

# Interpretación
if p_valor > 0.05:
    print("No se puede rechazar la hipótesis nula: los datos tienen una distribución normal.")
else:
    print("Se rechaza la hipótesis nula: los datos no tienen una distribución normal.")
```

Como el P- valor está muy por debajo de 0,05, se rechaza la hipotesis nula y muestra que los datos no muestran una distribución normal, aunque ya se podía confirmar mediante los datos de curtosis y gráficos.


## 2. Año

```python
df3=df['año']
#  Prueba de D'Agostino
stat, p_valor = normaltest(df3)

# Mostrar los resultados
print("Estadístico de prueba:", stat)
print("P-valor:", p_valor)

# Interpretación
if p_valor > 0.05:
    print("No se puede rechazar la hipótesis nula: los datos tienen una distribución normal.")
else:
    print("Se rechaza la hipótesis nula: los datos no tienen una distribución normal.")
```

Como el P- valor está muy por debajo de 0,05, se rechaza la hipotesis nula y muestra que los datos no muestran una distribución normal, aunque ya se podía confirmar mediante los datos de curtosis y gráficos.

### Comparar medianas. Prueba de Kruskal-Wallis

```python
# diferencia Nivel laboral y promedio de ganancias

# Filtrar los datos por niveles laborales (ocupacionales)
filtered_data = df4.dropna(subset=['prom_ganancias'])
grupos = filtered_data.groupby('nivel_laboral')['prom_ganancias']

# Crear listas con las ganancias para cada nivel laboral
niveles = [grupo for _, grupo in grupos]

# Realizar la prueba de Kruskal-Wallis
stat, p_value = kruskal(*niveles)

# Resultados
print(f'Estadístico H: {stat}')
print(f'p-valor: {p_value}')

if p_value < 0.05:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")
```

El  resultado de estadístico H (115.94) sugiere que existe diferencias entre los grupos debido a que es un valor bastante grande. 
Por otro lado el valor del P-valor es menor a 0,05 por lo que rechazamos la hipotesis nula y asumimos que existen diferencias significativas entre las diferentes ocupaciones

```python
#Diferencias en genero en el promedio de ganancias

# Agrupar los datos por sexo y extraer las ganancias promedio
grupos_sexo = filtered_data.groupby('sexo')['prom_ganancias']

# Crear listas con las ganancias para cada sexo
niveles_sexo = [grupo for _, grupo in grupos_sexo]

# Realizar la prueba de Kruskal-Wallis para comparar por sexo
stat_sexo, p_value_sexo = kruskal(*niveles_sexo)

# Resultados
stat_sexo, p_value_sexo
# Resultados
print(f'Estadístico H: {stat_sexo}')
print(f'p-valor: {p_value_sexo}')

if p_value_sexo < 0.05:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")
```

El valor Estadistico (4,86) indica un valor muy bajo por lo que no se puede observar grandes diferencias.
el Pvalor nos confirma que no existen diferencias estadisticamente hablado, debido a que 0,088 es mayor que 0,05 por lo que no rechazamos la hipotesis nula. Sin embargo, el valor es cercano a 0'05 por lo que podría sugerir una diferencia sútil.
